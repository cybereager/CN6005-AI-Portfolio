```python
# Step 1: Install required library (run once if not installed)
# !pip install mlxtend

# Step 2: Import libraries
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Step 3: Define the transaction dataset
transactions = [
    ['Mango', 'Onion', 'Nintendo', 'Key-chain', 'Eggs', 'Yo-yo'],
    ['Doll', 'Onion', 'Nintendo', 'Key-chain', 'Eggs', 'Yo-yo'],
    ['Mango', 'Apple', 'Key-chain', 'Eggs'],
    ['Mango', 'Umbrella', 'Corn', 'Key-chain', 'Yo-yo'],
    ['Corn', 'Onion', 'Key-chain', 'Ice-cream', 'Eggs']
]

# Step 4: Convert transactions to one-hot encoded format
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_array, columns=te.columns_)

print("One-hot encoded transaction data:")
print(df_encoded)

# Step 5: Apply Apriori algorithm (min_sup = 60%)
frequent_itemsets = apriori(
    df_encoded,
    min_support=0.6,
    use_colnames=True
)

print("\nFrequent Itemsets (min_support = 60%):")
print(frequent_itemsets)

# Step 6: Generate association rules (min_conf = 80%)
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.8
)

# Keep only meaningful rules
rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]

print("\nStrong Association Rules (min_conf = 80%):")
print(rules)
```
